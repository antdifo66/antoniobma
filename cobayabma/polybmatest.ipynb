```python
'''
Please run each cell individually.

**Between runs of this program, please DELETE the output folders it creates, which in this case should be 'chainstriple3',
'chainstriple4','chainstriple5','chainstriple6'.
Otherwise, the Cobaya MCMC chain sampling will not run and returns an error.
'''

'''
The purpose of this program is to test Bayesian Model Averaging in the M-closed case: the true data generator IS one
of the models chosen in BMA analysis.

The third cell runs through a loop, running the program with differing amounts of noise.
'''

'''
Dependencies required (from repo):
FastMPClinpara
polynomialbma
'''

import polynomialbma as pbma

atrue=0.034
btrue=-0.033
ctrue=-1.03

infopara, infolin, infocube = pbma.createfit(0.1, atrue, btrue, ctrue, 0)
```

```python
pbma.runmcmcs(infopara, infolin, infocube, atrue, btrue, ctrue, 0)  #don't forget to delete the folder
```

```python
import polynomialbma as pbma
for i in range(3):
    infopara, infolin, infocube = pbma.createfit(0.35*(i+1), atrue, btrue, ctrue, i+1)
    pbma.runmcmcs(infopara, infolin, infocube, atrue, btrue, ctrue, i+1)
    
```
